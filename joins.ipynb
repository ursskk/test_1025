```python
df_orders = spark.table('samples.tpch.orders')
df_customers = spark.table('samples.tpch.customer')
df_lineitem = spark.table('samples.tpch.lineitem')

df_orders.createOrReplaceTempView("orders")
df_customers.createOrReplaceTempView("customers")
df_lineitem.createOrReplaceTempView("lineitem")
```

```python
print(df_orders.columns)
print(df_customers.columns)
print(df_lineitem.columns)
```

```python
from pyspark.sql.functions import col
df_join1 = df_orders.alias("ord").join(df_customers.alias("cust"), col("ord.o_orderkey") == col("cust.c_custkey"), "inner")
df_join1.count()
```

```python
jointstDF = spark.sql("""
                      select count(*)
                      from orders ord
                      join customers cust on ord.o_orderkey = cust.c_custkey
                      where not exists (select 1 from lineitem li where ord.o_custkey = li.l_orderkey) 
                      """
                                           
                      )
display(jointstDF.show())                 
```

```python
df_1 = df_orders.alias('ord').join(df_customers.alias('cust'),col("ord.o_orderkey") == col("cust.c_custkey"),"inner")
df_final = df_1.alias("OC").join(df_lineitem.alias("li"), col("OC.o_custkey") == col("li.l_orderkey"),"anti")
df_final.count()
```
